## 1-Marche aléatoire 1D sans boucle

```python
import numpy as np # On importe numpy 
```

```python
def marche_aleatoire(p, N):
    # Tirages aléatoires pour N pas
    tirages = np.random.rand(N)
    # +1 si tirage < p, sinon -1
    pas = np.where(tirages < p, 1, -1)
    #Le déplacement final après N pas est la somme des +1 et -1
    return pas.sum()
```

 Exemple d utilisation

```python
print(marche_aleatoire(0.5, 10))  # p=0.5 , N=10
print(marche_aleatoire(0.7, 100)) # p=0.7 , N=100
print(marche_aleatoire(0.3, 50))  # p=0.3 , N=50
```

2-Monte-Carlo avec boucles 

```python
def monte_carlo(p, N, M):
    #un tableau vide de taille M pour stocker les déplacements finaux
    deplacements = np.zeros(M)
    # On répète M fois l’expérience
    for i in range(M):
        # On stocke le déplacement final
        deplacements[i] = marche_aleatoire(p, N)
    # La moyenne des déplacements sur M expériences
    moyenne = deplacements.mean()
    # L'écart-type 
    ecart_type = deplacements.std()
    # On renvoie la moyenne, l'écart-type 
    return moyenne, ecart_type
```

```python
p = 0.6   # probabilité de faire un pas +1
N = 10   # nombre de pas par marche
M = 10  # nombre d’expériences Monte Carlo
```

```python
moy, ec = monte_carlo(p, N, M)
```

Affichage des résultats

```python
print("Moyenne du déplacement :", moy)
print("Écart-type :", ec)
```

3-Variation de M

```python
p = 0.6  # probabilité de faire un pas +1
N = 100  # nombre de pas par marche
```

Les 4 valeurs de M 

```python
M_values = [10, 100, 1000, 10000]
```

Tableau résultat : 2 lignes et 4 colonnes

```python
resultats = np.zeros((2, 4))
```

remplit le tableau

```python
for i in range(len(M_values)):
    M = M_values[i]
    moy, ec = monte_carlo(p, N, M)
    resultats[0, i] = moy        # Ligne 1 = moyennes
    resultats[1, i] = ec         # Ligne 2 = écarts-types
```

Affichage du tableau final

```python
print("Tableau des moyennes et écarts-types :")
print(resultats)
```

Comparaison des valeurs théoriques

```python
for i, M in enumerate(M_values):
    print(f"M = {M} : moyenne sim = {resultats[0,i]:.2f}, ecart-type sim = {resultats[1,i]:.2f}")
```

```python
mu_theorique = N * (2*p - 1)
sigma_theorique = np.sqrt(N * 4*p*(1-p))
```

```python
for M in M_values:
    print(f"M = {M} : Moyenne théorique = {mu_theorique:.2f}, Écart-type théorique = {sigma_theorique:.2f}")
```
